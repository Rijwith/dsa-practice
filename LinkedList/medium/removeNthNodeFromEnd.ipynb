```python
def removeNthFromEnd(head, n):
    dummy = ListNode(0)
    dummy.next = head
    fast = slow = dummy

    # move fast ahead by n+1
    for _ in range(n + 1):
        fast = fast.next

    # move both pointers
    while fast:
        fast = fast.next
        slow = slow.next

    # remove nth node
    slow.next = slow.next.next
    return dummy.next
```
