```python
from collections import deque

def levelOrder(root):
    if not root:
        return []

    result = []
    q = deque([root])

    while q:
        level = []
        size = len(q)

        for _ in range(size):
            node = q.popleft()
            level.append(node.val)

            if node.left:
                q.append(node.left)
            if node.right:
                q.append(node.right)

        result.append(level)

    return result
```
