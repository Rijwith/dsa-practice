```python
class CircularQueue:
    def __init__(self, k):
        self.k = k
        self.q = [0] * k
        self.front = 0
        self.rear = 0
        self.count = 0

    def enqueue(self, value):
        if self.isFull():
            return False
        self.q[self.rear] = value
        self.rear = (self.rear + 1) % self.k
        self.count += 1
        return True

    def dequeue(self):
        if self.isEmpty():
            return False
        self.front = (self.front + 1) % self.k
        self.count -= 1
        return True

    def Front(self):
        if self.isEmpty():
            return None
        return self.q[self.front]

    def Rear(self):
        if self.isEmpty():
            return None
        return self.q[(self.rear - 1) % self.k]

    def isEmpty(self):
        return self.count == 0

    def isFull(self):
        return self.count == self.k
```
