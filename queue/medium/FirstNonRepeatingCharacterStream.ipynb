```python
from collections import deque

def firstNonRepeating(stream):
    freq = {}
    q = deque()
    result = []

    for ch in stream:
        freq[ch] = freq.get(ch, 0) + 1
        q.append(ch)

        while q and freq[q[0]] > 1:
            q.popleft()

        if q:
            result.append(q[0])
        else:
            result.append(-1)

    return result
```
