```python
from collections import deque

class MyStack:
    def __init__(self):
        self.q = deque()

    def push(self, x):
        self.q.append(x)
        # rotate
        for _ in range(len(self.q) - 1):
            self.q.append(self.q.popleft())

    def pop(self):
        if self.empty():
            return None
        return self.q.popleft()

    def top(self):
        if self.empty():
            return None
        return self.q[0]

    def empty(self):
        return len(self.q) == 0
```
