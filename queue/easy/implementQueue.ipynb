```python
class Queue:
    def __init__(self):
        self.q = []

    def enqueue(self, x):
        self.q.append(x)

    def dequeue(self):
        if self.isEmpty():
            return None
        return self.q.pop(0)

    def front(self):
        if self.isEmpty():
            return None
        return self.q[0]

    def isEmpty(self):
        return len(self.q) == 0
```
