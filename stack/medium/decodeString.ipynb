```python
def decodeString(s: str) -> str:
    countStack = []
    stringStack = []
    currStr = ""
    currNum = 0

    for ch in s:
        if ch.isdigit():
            currNum = currNum * 10 + int(ch)

        elif ch == '[':
            countStack.append(currNum)
            stringStack.append(currStr)
            currNum = 0
            currStr = ""

        elif ch == ']':
            repeat = countStack.pop()
            prevStr = stringStack.pop()
            currStr = prevStr + currStr * repeat

        else:
            currStr += ch

    return currStr
```
