```python
def removeKdigits(num: str, k: int) -> str:
    stack = []

    for digit in num:
        while stack and k > 0 and stack[-1] > digit:
            stack.pop()
            k -= 1
        stack.append(digit)

    # If k still > 0, remove from end
    while k > 0:
        stack.pop()
        k -= 1

    # Remove leading zeros
    result = ''.join(stack).lstrip('0')

    return result if result else "0"
```
