```python
def largestRectangleArea(heights):
    stack = []
    maxArea = 0
    n = len(heights)

    for i in range(n + 1):
        curr_height = heights[i] if i < n else 0

        while stack and curr_height < heights[stack[-1]]:
            h = heights[stack.pop()]
            width = i if not stack else i - stack[-1] - 1
            maxArea = max(maxArea, h * width)

        stack.append(i)

    return maxArea
```
