```python

def isValid(s):
    """
    :type s: str
    :rtype: bool
    """
    dict1={']':'[','}':'{',')':'('}
    lst1=[]
    for i in s:
        if i in dict1:
            if dict1[i] in lst1 and dict1[i]==lst1[-1]:
                lst1.pop()
            else:
                return False
        else:
            lst1.append(i)
    if len(lst1)!=0:
        return False
    else:
        return True
```

```python
s = "()[]{}"
isValid(s)
```
